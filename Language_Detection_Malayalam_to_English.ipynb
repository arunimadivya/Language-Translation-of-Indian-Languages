**Tokenizing**

```python
from keras.preprocessing.text import Tokenizer
def tokenize(x):
    #Initializing the tokenizer
    tokenizer = Tokenizer()
    #Updates vocabulary based on the input texts
    tokenizer.fit_on_texts(x)
    return tokenizer.texts_to_sequences(x), tokenizer
```

**Padding**

```python
from keras.preprocessing.sequence import pad_sequences
def pad(text, length=None):
    #Pad the text to the length mentioned after each sequence
    return pad_sequences(text, maxlen=length, padding='post')
```

**Logits to Text**

The below function will bridge the gab between the logits from the neural network to the Malayalam translation.

```python
def logitsToText(logits, tokenizer):
    indexToWords = {id: word for word, id in tokenizer.word_index.items()}
    #Join the words with higher prediction
    return ' '.join([indexToWords[prediction] for prediction in np.argmax(logits, 1) if prediction!=np.int64(0)])
```

**RNN**

```python
from keras.layers import GRU, Input, Dense, TimeDistributed
from keras.models import Model, Sequential
from keras.layers import Activation
from tensorflow.keras.optimizers import Adam
from keras.losses import sparse_categorical_crossentropy


def simple_model(inputShape, outputSeqLen, malayalamVocabSize, englishVocabSize, learning_rate=0.1):
    model = Sequential()
    #3D tensor of shape (batch_size, timesteps, 256)
    model.add(GRU(128, dropout=0.1,input_shape=inputShape[1:], return_sequences=True) )
    model.add(Dense(englishVocabSize, activation='softmax') )
    model.summary()
    
    model.compile(loss=sparse_categorical_crossentropy,
                  optimizer=Adam(learning_rate),
                  metrics=['accuracy'])
    return model
```

**LSTM with an embedding layer**

```python
from keras.layers.embeddings import Embedding
from keras.layers import LSTM
from keras.layers import GRU, Input, Dense, TimeDistributed
from keras.models import Model, Sequential
from keras.layers import Activation
from tensorflow.keras.optimizers import Adam
from keras.losses import sparse_categorical_crossentropy

def LSTMandEmbedModel(inputShape, outputSeqLen, malayalamVocabSize, englishVocabSize, learning_rate=0.1):
    model = Sequential()
    model.add(Embedding(max(malayalamVocabSize, englishVocabSize) ,128 , input_length=outputSeqLen))
    model.add(LSTM(128, dropout=0.1, return_sequences=True))
    model.add(Dense(englishVocabSize, activation='softmax'))
    model.summary()
    model.compile(loss=sparse_categorical_crossentropy,
                  optimizer=Adam(learning_rate),
                  metrics=['accuracy'])
    
    return model
```

```python
embedLSTMModel = LSTMandEmbedModel(
    malayalamTextNew.shape,
    malayalamTextNew.shape[1],
    len(malayalamTextTokenizer.word_index)+1,
    len(englishTextTokenizer.word_index)+1)
if os.path.exists(os.path.join("model", "LSTM.h5"))== False:
    embedrnn = embedLSTMModel.fit(malayalamTextNew, englishTextNew, batch_size=100, epochs=20, validation_split=0.2)
else:
    embedrnn = load_model(os.path.join("model", "LSTM.h5"))
```

**Bidirectional LSTM with embedding layer**

```python
from keras.layers import Bidirectional
def biLSTMModel(inputShape, outputSeqLen, malayalamVocabSize, englishVocabSize, learning_rate=0.1):
    model = Sequential()
    model.add(Embedding(max(malayalamVocabSize, englishVocabSize) ,128 , input_length=outputSeqLen))
    model.add(Bidirectional(LSTM(128, return_sequences=True, dropout=0.1),
                        input_shape=inputShape[1:]))
    
    model.add(TimeDistributed(Dense(englishVocabSize, activation='softmax') ))
    model.summary() 
    model.compile(loss=sparse_categorical_crossentropy,
                  optimizer=Adam(learning_rate),
                  metrics=['accuracy'])  
    return model
```

```python
bidirLSTMModel = biLSTMModel(
    malayalamTextNew.shape,
    englishTextNew.shape[1],
    len(malayalamTextTokenizer.word_index)+1,
    len(englishTextTokenizer.word_index)+1)

if os.path.exists(os.path.join("model", "biRnn.h5"))== False:
    biDirLSTMModel = bidirLSTMModel.fit(malayalamTextNew, englishTextNew, batch_size=100, epochs=20 ,validation_split=0.2)
else:
    bidirLSTMModel = load_model(os.path.join("model", "biRnn.h5"))
```

```python
score = bidirLSTMModel.evaluate(malayalamTextNew, englishTextNew, verbose=0)
print("Train accurancy: ", score[1])
```
